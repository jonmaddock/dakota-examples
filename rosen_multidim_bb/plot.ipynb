# Plot Rosenbrock multidimensional parameter study

```python
import pandas as pd
from pathlib import Path
import numpy as np
from matplotlib import pyplot as plt
import matplotlib
```

```python
data_file = Path("rosen_multidim.dat")
data = pd.read_csv(data_file, sep="\s+")
# Replace 0s in response with small postive values for log scale to work
data["response_fn_1"] = data["response_fn_1"].replace(0,1e-6)
data
```

```python
fig, ax = plt.subplots()
tcf = ax.tricontourf(data["x1"], data["x2"], data["response_fn_1"], norm="log", locator=matplotlib.ticker.LogLocator())
ax.axis("scaled")
fig.colorbar(tcf, label="Value of Rosenbrock function")
ax.set_title("Rosenbrock function parameter study in Dakota")
ax.set_xlabel("x")
ax.set_ylabel("y")

# Plot actual sample points
ax.scatter(data["x1"], data["x2"], marker=".", label="Sample points", s=10)
ax.legend()
```
