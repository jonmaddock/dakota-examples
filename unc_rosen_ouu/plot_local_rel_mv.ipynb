# Plot Uncertain Rosenbrock Local Reliability Study

A reliability analysis is performed at the solution, safe (low constraint sum region) and dangerous (high constraint sum region) points (see con_space_complex notebook). An epistemic uncertainty has been applied to both constraints.

```python
import pandas as pd
from pathlib import Path
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
# Read in output file
output_filename = Path("responses_stdout_sol.txt")
with open(output_filename, "r") as output_file:
    for line_no, line_str in enumerate(output_file.readlines()):
        if "Cumulative Distribution Function (CDF)" in line_str:
            start = line_no + 3
        elif "Iterator local_reliability completed" in line_str:
            finish = line_no - 2
    
cdf_data = np.loadtxt(output_filename, skiprows=start, max_rows=(finish-start))
# cdf_data
fig, ax = plt.subplots()
ax.plot(cdf_data[:,0], cdf_data[:,1], marker="o")
ax.set_xlabel("Response (sum of constraints)")
ax.set_ylabel("CDF")
ax.set_title("Local reliability at solution")
```

```python
# Read in output file
output_filename = Path("responses_stdout_safe.txt")
with open(output_filename, "r") as output_file:
    for line_no, line_str in enumerate(output_file.readlines()):
        if "Cumulative Distribution Function (CDF)" in line_str:
            start = line_no + 3
        elif "Iterator local_reliability completed" in line_str:
            finish = line_no - 2
    
cdf_data = np.loadtxt(output_filename, skiprows=start, max_rows=(finish-start))
# cdf_data
fig, ax = plt.subplots()
ax.plot(cdf_data[:,0], cdf_data[:,1], marker="o")
ax.set_xlabel("Response (sum of constraints)")
ax.set_ylabel("CDF")
ax.set_title("Local reliability at safe point")
```
