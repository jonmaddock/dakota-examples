```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
```

```python
fig, ax = plt.subplots()
a = np.random.rand(4)
b = np.random.rand(4)
tcf = ax.tricontourf(a, b, a)
```

```python
a = np.array([1,2,3]*3)
b = np.array([1,1,1,2,2,2,3,3,3])
c = b*2
data = np.array([a, b, c]).T
data
df = pd.DataFrame(data, columns=["a", "b", "c"])
df
```

```python
fig, ax = plt.subplots()
tcf = ax.tricontourf(a, b, c)
fig.colorbar(tcf, label="c")
```

```python
# Create one -to-many mapping: same value of (a,b) can have multiple values of c
a = np.array([1,2,3]*3)
a = np.concatenate((a,a))
b = np.array([1,1,1,2,2,2,3,3,3])
b = np.concatenate((b,b))
c = np.concatenate((np.array([1]*9), np.array([2]*9)))
data = np.array([a, b, c]).T
data
df = pd.DataFrame(data, columns=["a", "b", "c"])
df
```

```python
fig, ax = plt.subplots()
tcf = ax.tricontourf(a, b, c)
fig.colorbar(tcf, label="c")
```
